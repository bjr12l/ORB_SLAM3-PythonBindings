```python
import cv2
import os
import ffmpeg

from pathlib import Path
```

```python
VIDEO_DIR = Path("../data/copter_flir/")
MKV = VIDEO_DIR / "mkv"
MP4 = VIDEO_DIR / "mp4"
```

```python
# Get all .mp4 files from the specified directory
video_files = sorted(os.listdir(MKV))
```

```python
for video_file in video_files:
    # Construct the output file name by replacing .mkv with .mp4
    output_file = video_file.rsplit(".", 1)[0] + ".mp4"

    # Convert the video
    ffmpeg.input(str(MKV / video_file)).output(
        str(MP4 / output_file), 
        vcodec='libx264', acodec='aac'
    ).run()
```

```python
# Get all .mp4 files from the specified directory
video_files = sorted(os.listdir(MP4))
```

```python
# Get properties of the video using the first video file
video = cv2.VideoCapture(str(MP4 / video_files[0]))
```

```python
frame_rate = video.get(cv2.CAP_PROP_FPS)
frame_width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
```

```python
# Create a VideoWriter object
out = cv2.VideoWriter(str(VIDEO_DIR / "combined_video.mp4"), cv2.VideoWriter_fourcc(*'mp4v'), frame_rate, (frame_width, frame_height))

for video_file in video_files:
    video = cv2.VideoCapture(str(MP4 / video_file))
    while video.isOpened():
        ret, frame = video.read()
        if not ret:
            break
        out.write(frame)
    video.release()

out.release()
```
